# Sentiment Analysis on Covid-19 vaccines in Ireland
## Exploration of tweets after labelling process
### Just for global tweets (tweets coming form batch covid_vaccine_global)

```python
# Data Manipulation
import pandas as pd
import numpy as np
import string, re, datetime

# Data Visualization
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

from wordcloud import WordCloud
from sklearn.feature_extraction.text import CountVectorizer

from matplotlib.cbook import boxplot_stats 

# Own modules
from layer_data_access import tweet_data
from layer_classes import my_yaml
```

```python
# Get token excluding stop words and keywords
access_data = tweet_data.tweet_data()
tweets_df = access_data.get_tweets_train_test()
del access_data
```

```python
# Group batches, including conversations and replies
sentiment = tweets_df.label.value_counts()

# Plot results
plt.figure(figsize=(8,5), dpi=100)
plt.rcParams.update({'font.size': 13})
plt.pie(x = sentiment.to_list(), autopct = '%1.1f%%', shadow = True, colors=['tab:green', 'tab:orange', 'tab:red'])
plt.axis('equal')
plt.title("Distribution of sentiment\n in global tweets")
plt.legend(sentiment.index.tolist(), loc=(0.8,0.7), title = 'Sentiment')
plt.tight_layout()
plt.show()
```
